## Как создавать Merge Request в публичный репозиторий

Да, этот формат будет работать в VS Code, потому что я не знаю, как заставить его работать в других приложениях. 

##### 1. Клонируем репозиторий

Сначала нужно создать локальную копию репозитория. Сначала создайте пустую папку которую вы отроете в VSC, удостоверьесь что в VSC вы подлюченны к своему github аккаунту. 

```bash
git clone github.com/LocalDoc/helper-bot-
```
---

###### 2. Создаём свою ветку

Не работайте прямо в `main` или `master` пожалуйста. Создайте новую ветку для своих изменений: 

```bash
git checkout -b your-feature-branch
```

Название ветки пусть отражает вашу задачу или исправление, ну или хоть папку которую вы меняете. (your-feature-branch - имя которое вы подбираете именно вашему бранчу) 

##### 3. Вносим изменения

Редактируйте, добавляйте или удаляйте файлы.

##### 4. Сохраняем изменения (commit)

Сначала добавим файлы в staging (можно тоже сделать это плюсиком в части source control vsc): 

```bash
git add .
```
Затем создаём коммит:

```bash
git commit -m "Краткое описание того, что делает этот коммит."
```

Всегда **описывайте, что вы коммитите**, чтобы мы все понимали цель изменений.

##### 5. Отправляем ветку на GitHub

```bash
git push origin your-feature-branch
```

###### 6. Создаём Merge Request (Pull Request)

1. Откройте репозиторий на GitHub в браузере.
2. Нажмите **Compare & pull request** рядом с вашей веткой.
4. Укажите целевую ветку (`main` должно быть).
5. Отправьте pull request.

###### 7. Notes:
Всем выдались права коллаборатора если в описанном тут процессе
